## 获取股票数据




<font size =4 color = "blue" ><b>示例 ： </b> 获取A股的格力电器的历史行情数据，不复权 </font>

```python
# 所有历史数据
df = ak.stock_zh_a_daily(symbol="sz000651", factor="") #sh600000  ,sz000651
df
```

<font size =4  ><b>获取最近20个交易日数据 </b></font>

```python
df.tail(20)
```

<font size =4  ><b>获取最近1年的交易数据 </b></font>

当前日期 ： 2020-03-20

```python
today = time.strftime('%Y-%m-%d' , time.localtime())
today
```

```python
startdate = (datetime.strptime(today,"%Y-%m-%d")-timedelta(days =365)).strftime("%Y-%m-%d")
startdate
```

```python
df.loc[df.index>startdate,:]
```

## 获取指数数据

```python
stock_df = ak.stock_zh_index_spot()
stock_df
```

```python
stock_zh_index_daily_df = ak.stock_zh_index_daily(symbol="sz399996")
stock_zh_index_daily_df
```

## 可视化

```python
# 设参数
x1 = stock_zh_index_daily_df.index 
y1 = stock_zh_index_daily_df["close"]
```

```python
trace = go.Scatter(
     x = x1,
     y = y1     
)
fig = dict(data = [trace] , layout = dict(showlegend=False))
py.iplot(fig)
```
